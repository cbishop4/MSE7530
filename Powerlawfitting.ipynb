# Experimenting with power law fits

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
qs = np.linspace(0.001,0.01,10001)
```

```python
def powerbg(q,constant):
    bg = q**constant
    return bg
```

### Experimentation

```python
trybg = powerbg(qs,-3.)
```

```python
plt.plot(qs,trybg)
```
