Pandas loads files into **"dataframes"**. This is the main structure in Pandas. As Pandas is another one of those highly used Python packages, the documentation is really nice. Check out their main page: https://pandas.pydata.org/docs/index.html and use the "Getting Started" and "User guide" to learn the main concepts, working with dataframes, and find many tutorials.

The above was an easy import, because the header line was the first line. Before I use Colab's special dataframe functions, I'll show you how I would plot this in the olden days.

```python
fig, ax = plt.subplots()
ax.plot(loaded['XrayEnergy (eV)'],loaded['Absorption'],'-o',markersize=2)
ax.set_xlabel('X-ray Energy (eV)'); ax.set_ylabel('Absorption')
ax.set_title('NEXAFS spectrum of HOPG at 30 $\\degree$ incidence')
```
